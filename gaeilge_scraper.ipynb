```python
#####################
## DATA EXTRACTION ##
#####################

import requests
from bs4 import BeautifulSoup
import re


def get_page(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    return soup

    
soup = get_page("https://songsinirish.com/tags/sean-nos/")
```

```python
# build link list
links = []
save = 0

for link in soup.findAll("a"):
    # start saving links once at A-Z index
    if "A bhean úd thíos" in link:
        save = 1
    
    if save == 1:
        links.append(link.get('href'))
    
    if "Úrchnoc Chéin Mhic Cáinte" in link:
        save = 0
        
print(len(links))
```

```python
# crawl links and scrape text
gaeilge = []

for link in links:
    print('Link: ', link)
    soup = get_page(link)
    
    for text in soup.findAll("script"):
        if "teacs-gaeilge" in str(text):
            gaeilge_text = re.findall(r'(?<=var htmlString = \")(.*)(?=\";)', str(text))
            # remove tags
            gaeilge_clean = re.sub("<[^>]*>", " ", str(gaeilge_text))
            gaeilge.append(gaeilge_clean)
    
print(len(gaeilge))
```

```python
## START HERE IF YOU HAVE ALREADY SCRAPED THE DATA
import csv
import pandas as pd
import numpy as np
from nltk.tokenize import RegexpTokenizer
import nltk

# export to csv
#with open('gaeilge_scrape.csv', 'w', newline='', encoding='UTF-8') as f:
#    writer = csv.writer(f)
#    writer.writerows(gaeilge)
    
# import file
gaeilge = []

with open('gaeilge_scrape.csv', 'r', newline='', encoding='UTF-8') as f:
    reader = csv.reader(f)
    for row in reader:
        gaeilge.append(''.join(row))
        
print(gaeilge[0])
print(len(gaeilge))
```

```python
###################
## DATA CLEANING ##
###################
# create gaeilge stopwords
gaeilge_stopwords = ['deas', 'raibh', 'nuair','dh', 'mb', 'c', 'e', 'it', 'll', 'mh', 'be', 'at', 'ea', 'he', 'all', 'h', 'm', 'd', 'n', 'b', 't', 'se', 'ne', 'ab', 'g', 'ín', 'ait', 'aí', 'hí', 'cha', 'air', 'os', 'é', 'fá', 'on', 'om', 'id', 'ga', 'hé', 'sea', 'o', 'á', 'as', 'am', 'ó', 'ba', 'na', 'im', 'í', 's', 'a', 'nach', 'gan', 'riamh', 'ariamh', 'chuaigh', 'chuile', 'amuigh', 'dimigh', 'liomsa', 'bhíodh', 'mbíonn', 'ndiaidh', 'istigh', 'cheann', 'thaobh', 'mbeadh', 'bheadh', 'bheith', 'is', 'nó', 'eile', 'idir', 'astu', 'thíos', 'as', 'ár', 'táim', 'bhí', 'aisti', 'chugat', 'orainn', 'anseo', 'déanamh', 'uaithi', 'ná', 'aon', 'suas', 'céanna', 'fúm', 'cathain', 'má', 'bhíothas', 'agaoibh', 'asainn', 'tá', 'chomh', 'tuilleadh', 'fút', 'a', 'éigin', 'freisin', 'ionat', 'fúthu', 'sé', 'muid', 'sna', 'ort', 'níl', 'chuainn', 'tamall', 'níos', 'seo', 'uait', 'beimid', 'leas', 'féin', 'chucu', 'cad', 'libh', 'sinn', 'araon', 'uirthi', 'iad', 'uaim', 'mar', 'beith', 'agam', 'uainn', 'faoi', 'chugam', 'cé', 'de', 'úire', 'trí', 'asat', 'fosta', 'ní', 'ar', 'ionam', 'cá', 'dtí', 'go', 'dhéanamh', 'aici', 'gcarr', 'beidh', 'orm', 'tríd', 'tamaill', 'mise', 'chuige', 'thuas', 'mó', 'leat', 'iontu', 'agat', 'do', 'bhfuil', 'ach', 'cár', 'timpeall', 'taobh', 'inti', 'táthar', 'bhíomar', 'sa', 'anois', 'againn', 'siad', 'chuici', 'táimid', 'fúinn', 'le', 'agus', 'dom', 'isteach', 'i', 'gach', 'cén', 'ansin', 'ann', 'tú', 'leis', 'uair', 'tosach', 'fúibh', 'uaidh', 'amach', 'orthu', 'di', 'duit', 'ionainn', 'in', 'arís', 'siúd', 'bhrí', 'leo', 'é', 'uathu', 'ó', 'carr', 'tusa', 'aige', 'dhiaidh', 'na', 'dóibh', 'an', 'linn', 'síos', 'charr', 'aice', 'acu', 'thar', 'sí', 'léi', 'liom', 'mé', 'mo', 'déanaim', 'chaoi', 'asam', 'dó', 'conas', 'dúinn', 'curfá', 'cúrfa', 'tá', 'ag']

# lowercase, no punctuation, no stop words
gaeilge_clean = []

for song in gaeilge:
    song = song.lower()
    tokenizer = RegexpTokenizer(r'\w+')
    punc_free = tokenizer.tokenize(song)
    stop_free = ' '.join([word for word in punc_free if word not in gaeilge_stopwords])
    gaeilge_clean.append(stop_free)

print(len(gaeilge_clean))
print(gaeilge_clean[0])
```

```python
# combine all songs
gaeilge_full = ''.join(gaeilge_clean)
print(len(gaeilge_full), ' total characters in corpus')
    
# word tokenize
gaeilge_words = nltk.word_tokenize(gaeilge_full)
print(len(gaeilge_words), ' total words in corpus')

# frequency distribution
dist = nltk.FreqDist(gaeilge_words)
print(len(dist.keys()), ' total unique words in corpus')

# create frequency df
freq_df = pd.DataFrame(columns=['word', 'frequency'])

for word in dist.keys():
    freq_df = freq_df.append({'word': word, 'frequency': dist[word]}, ignore_index=True)

# index frequency df by word
word_df = freq_df.set_index('word')

# words in Irish can have an initial mutation, search for these mutations and replace with nominative form
# Examples:
# cailín = chailín
# abhaile = bhaile, mbaile
# caith = chaith
# máithrín = mháithrín

mutations = ['mb','dt','gc', 'nd', 'ng', 'bp']

for word in word_df.index:
    nom_word = np.NaN
    
    if len(word) >= 2:
        if word[1] == 'h' or word[:2] in mutations or word[:3] == 'bhf':
            if word[1] == 'h':
                nom_word = ''.join([word[:1],word[2:]])
            elif word[:2] in mutations:
                nom_word = word[1:]
            elif word[:3] == 'bhf':
                nom_word = word[2:]

        if not pd.isnull(nom_word):
            if nom_word in word_df.index:
                # add non-nominative forms to nominative word
                word_df.loc[nom_word,'frequency'] = word_df.loc[nom_word,'frequency'] + word_df.loc[word,'frequency']
                # drop non-nominative form
                word_df = word_df.drop(word)
            
print(word_df.head())
print(word_df.shape)
```

```python
########################
## DATA VISUALIZATION ##
########################

%matplotlib notebook

import matplotlib.pyplot as plt

# create frequent words df
freqwords = [word for word in word_df.index if len(word) > 3 and word_df['frequency'].loc[word] > 150]

freq_word_df = pd.DataFrame(columns=['word','frequency'])

for word in freqwords:
    freq_word_df = freq_word_df.append({'word': word, 'frequency': word_df.loc[word,'frequency']}, ignore_index=True)

# sort by most frequent to least frequent
freq_word_df = freq_word_df.sort_values(by='frequency')
most_freq = freq_word_df['frequency'].max()

fig, ax = plt.subplots()
hbars = ax.barh(freq_word_df['word'], freq_word_df['frequency'])
ax.set_xlabel('Frequency')
ax.set_yticklabels(freq_word_df['word'])
ax.bar_label(hbars, labels=freq_word_df['frequency'])
ax.set_xlim(right=most_freq + 25) # leave space for labels
ax.set_title('Frequent Words in Sean-Nós Songs on SongsInIrish.com')
plt.show()
plt.savefig('freqwords.png')
```

```python
# words shared between songs
shared_df = pd.DataFrame()

for word in word_df.index:
    word_count = 0
    total_word_freq = 0
    
    for song in gaeilge_clean:
        word_freq = song.count(word)
        if word_freq > 0:
            word_count = word_count + 1
            total_word_freq = total_word_freq + word_freq
    
    shared_df = shared_df.append({'word': word, 'total songs word appears in': word_count, 'total times word appears': total_word_freq}, ignore_index=True)
    
print(shared_df.head())
print(shared_df.shape)
```

```python
most_shared_df = shared_df[shared_df['total songs word appears in'] > 100]
most_shared_df = most_shared_df[most_shared_df['word'].str.len() > 4]
print(most_shared_df.shape)

# sort by most shared to least shared
most_shared_df = most_shared_df.sort_values(by='total songs word appears in')
most_freq = most_shared_df['total times word appears'].max()

fig, ax = plt.subplots()
hbars = ax.barh(most_shared_df['word'], most_shared_df['total songs word appears in'])
ax.set_xlabel('Total Shared Songs')
ax.set_yticklabels(most_shared_df['word'])
ax.bar_label(hbars, labels=most_shared_df['total songs word appears in'])
ax.set_xlim(right=most_freq + 15) # leave space for labels
ax.set_title('Shared Words in Sean-Nós Songs on SongsInIrish.com')
plt.show()
plt.savefig('sharedwords.png')
```

```python
# frequent word use per song
shared_df['average word use per song'] = shared_df['total times word appears'] / shared_df['total songs word appears in']

repeat_df = shared_df[shared_df['average word use per song'] > 10]
repeat_df = repeat_df[repeat_df['word'].str.len() > 3]
print(repeat_df.shape)

# sort by most freq to least freq
repeat_df = repeat_df.sort_values(by='average word use per song')
most_freq = repeat_df['average word use per song'].max()

fig, ax = plt.subplots()
hbars = ax.barh(repeat_df['word'], repeat_df['average word use per song'])
ax.set_xlabel('Average Word Use Per Song')
ax.set_yticklabels(repeat_df['word'])
ax.bar_label(hbars, labels=repeat_df['average word use per song'])
ax.set_xlim(right=most_freq + 15) # leave space for labels
ax.set_title('Average Repetition in Sean-Nós Songs on SongsInIrish.com')
plt.show()
plt.savefig('repeatwords.png')
```
